```python
def poly_df(df):
    pdf = df
    
    columns = df.columns
    
    for col in columns:
        current_col = df[col]
        for feature in columns:
            feature_col = df[feature]
            pdf[f'{current_col.name}_X_{feature_col.name}'] = current_col * feature_col
            
    return pdf
```
