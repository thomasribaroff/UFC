#### Goal
The idea in this nb is to identify a list of columns that can be used as target variables, for building our later model, and to also create dummy cols for the categorial vars (e.g. stance being the most obvious one). 
There is one important consideration re splitting the data, in that, if we wanted to predict the win of either side, we'd need to do an additional split of the data. 
OR!! If we wanted to have double the amount of data!!!! we could split by fighter and have an additional row for each one with the fight stats for each one. This way we would be able to do a better prediction of individual fighter behaviour!!!!


Additional note: we actually have the vars for how much the opponent struck the R/B. Arguably we could just drop all those columns.
